```python
import torch
import yaml
from pathlib import Path
import numpy as np
from sklearn.metrics import accuracy_score, precision_recall_fscore_support
from tqdm.auto import tqdm

# Load your class names from the dataset YAML
with open('/content/drive/MyDrive/DL-PROJ/datasets/coco-2017-10k/dataset.yaml', 'r') as f:
    data = yaml.safe_load(f)  # Use safe_load to avoid using deprecated load
class_names = data['names']

# Initialize models
model_custom = torch.hub.load('ultralytics/yolov5', 'custom', path='/content/drive/MyDrive/DL-PROJ/yolov5_training/my_longer_run/weights/best.pt', force_reload=True)
models = {
    'custom': model_custom,
    'yolov5s': torch.hub.load('ultralytics/yolov5', 'yolov5s', force_reload=True),
    'yolov5m': torch.hub.load('ultralytics/yolov5', 'yolov5m', force_reload=True),
    'yolov5l': torch.hub.load('ultralytics/yolov5', 'yolov5l', force_reload=True),
    'yolov5x': torch.hub.load('ultralytics/yolov5', 'yolov5x', force_reload=True)
}

# Function to get true labels from the corresponding annotation file
def get_true_labels(img_path, labels_dir):
    label_path = labels_dir / f"{img_path.stem}.txt"
    if not label_path.exists():
        return []  # No annotations for this image
    with open(label_path) as f:
        labels = [int(line.split()[0]) for line in f.readlines()]
    return labels

val_images_dir = Path('/content/drive/MyDrive/DL-PROJ/datasets/coco-2017-10k/validation/yolo_labels/images/validation')
val_labels_dir = Path('/content/drive/MyDrive/DL-PROJ/datasets/coco-2017-10k/validation/yolo_labels/labels/validation')
image_paths = list(val_images_dir.glob('*.jpg'))

metrics = {'model_name': [], 'accuracy': [], 'precision': [], 'recall': [], 'f1_score': []}

for name, model in models.items():
    print(f'Evaluating {name}')
    y_true = []
    y_pred = []

    for img_path in tqdm(image_paths):
        true_labels = get_true_labels(img_path, val_labels_dir)
        if not true_labels:  # If there are no true labels for this image, skip it
            continue

        # Assume a single class per image for simplicity
        y_true.append(true_labels[0] if true_labels else -1)  # Use -1 for no detection

        results = model(str(img_path))
        pred_classes = results.pred[0][:, -1].cpu().numpy()
        y_pred.append(pred_classes[0] if len(pred_classes) > 0 else -1)  # Use -1 for no detection

    # Compute metrics for this model
    accuracy = accuracy_score(y_true, y_pred)
    precision, recall, f1, _ = precision_recall_fscore_support(y_true, y_pred, average='macro', zero_division=0)

    metrics['model_name'].append(name)
    metrics['accuracy'].append(accuracy)
    metrics['precision'].append(precision)
    metrics['recall'].append(recall)
    metrics['f1_score'].append(f1)

# Print the summarized metrics
for i in range(len(metrics['model_name'])):
    print(f"Results for {metrics['model_name'][i]}:")
    print(f"Accuracy: {metrics['accuracy'][i]}")
    print(f"Precision: {metrics['precision'][i]}")
    print(f"Recall: {metrics['recall'][i]}")
    print(f"F1 Score: {metrics['f1_score'][i]}\n")
```

```python
import torch
from pathlib import Path
import numpy as np
from tqdm.auto import tqdm
from sklearn.metrics import precision_recall_fscore_support

# Load class names and models
class_names = ['person', 'bicycle', 'car', 'motorcycle', 'airplane', 'bus', 'train', 'truck', 'boat', 'traffic light']  # example, adjust as per your dataset
classes_of_interest = {'car': 3, 'bus': 6, 'truck': 8, 'motorcycle': 4}

models = {
    'custom': torch.hub.load('ultralytics/yolov5', 'custom', path='path/to/custom/model.pt', source='local'),
    'yolov5s': torch.hub.load('ultralytics/yolov5', 'yolov5s', pretrained=True),
    'yolov5m': torch.hub.load('ultralytics/yolov5', 'yolov5m', pretrained=True),
    'yolov5l': torch.hub.load('ultralytics/yolov5', 'yolov5l', pretrained=True),
    'yolov5x': torch.hub.load('ultralytics/yolov5', 'yolov5x', pretrained=True),
}

# Define your dataset paths
image_dir = Path('/content/drive/MyDrive/DL-PROJ/datasets/coco-2014/yolo_labels/images/val')
label_dir = Path('/content/drive/MyDrive/DL-PROJ/datasets/coco-2014/yolo_labels/labels/val')
image_paths = list(val_images_dir.glob('*.jpg'))
def get_ground_truth_labels(image_path, label_dir, class_names, classes_of_interest):
    label_path = label_dir / (image_path.stem + '.txt')
    ground_truths = []
    if label_path.exists():
        with label_path.open('r') as f:
            for line in f.readlines():
                parts = line.strip().split()
                class_id = int(parts[0])
                if class_names[class_id] in classes_of_interest:
                    ground_truths.append(class_names[class_id])
    return ground_truths

# Evaluation
metrics = {}

for model_name, model in models.items():
    print(f"Evaluating {model_name}")
    all_gts, all_preds = [], []
    image_paths = list(image_dir.glob('*.jpg'))  # Or '*.png'

    for image_path in tqdm(image_paths):
        gts = get_ground_truth_labels(image_path, label_dir, class_names, classes_of_interest)
        all_gts.extend(gts)

        results = model(image_path)
        preds = results.pred[0]

        predicted_classes = []
        for pred in preds:
            conf, class_id = pred[4], int(pred[5])
            class_name = class_names[class_id]
            if class_name in classes_of_interest:
                predicted_classes.append(class_name)
        all_preds.extend(predicted_classes)

    precision, recall, f1, _ = precision_recall_fscore_support(all_gts, all_preds, labels=list(classes_of_interest.keys()), average='weighted')
    metrics[model_name] = {
        'precision': precision,
        'recall': recall,
        'f1': f1
    }

# Display metrics
for model_name, metric in metrics.items():
    print(f"{model_name} - Precision: {metric['precision']:.3f}, Recall: {metric['recall']:.3f}, F1: {metric['f1']:.3f}")
```
